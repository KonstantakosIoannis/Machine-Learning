```python
fig, axs = matplotlib.pyplot.subplots(1, 2, figsize=(10, 5))
axs[0].imshow(cat2)
axs[1].imshow(cat1)
axs[0].axis('off')
axs[1].axis('off')
axs[0].set_title('CV2 Image')
axs[1].set_title('Matplotlib Image')
```

```python
dog2=matplotlib.pyplot.imread(dogTrain[9])
dog3=cv2.imread(dogTrain[9])
```

```python
fig, axs = matplotlib.pyplot.subplots(1, 2, figsize=(10, 5))
axs[0].imshow(dog3)
axs[1].imshow(dog2)
axs[0].axis('off')
axs[1].axis('off')
axs[0].set_title('CV2 Image')
axs[1].set_title('Matplotlib Image')
```

```python
#the reason for this differences is because matplotlib is using the RGB protocol
#and the CV2 is using BGR
#thats why the cv2 image is bluer    :D
```

```python
matplotlib.pyplot.imshow(dog3)
```

```python
#of course we can change this protocol :
dog3 = cv2.cvtColor(dog3, cv2.COLOR_BGR2RGB)
```

```python
matplotlib.pyplot.imshow(dog3)
```

```python
dog4=matplotlib.pyplot.imread(dogTrain[42])
matplotlib.pyplot.imshow(dog4)
```
