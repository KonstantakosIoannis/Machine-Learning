### RECOMENDATION SYSTEM

```python
ratings  = pandas.read_csv('ratings.csv')
movies = pandas.read_csv('movies.csv')
```

```python
ratings = pandas.merge(ratings , movies , on='movieId')
```

```python
ratings.drop(['timestamp','genres'],axis=1,inplace=True) 
```

```python
ratings
```

```python
ratings.isna().sum()
```

```python
mydata = pandas.DataFrame(ratings.groupby('title')['rating'].mean())
mydata['count']=pandas.DataFrame(ratings.groupby('title')['rating'].count())
```

```python
mydata.sort_values(by='count',ascending=False)
```

```python
mytable = ratings.pivot_table(index='userId',columns='title',values='rating')
```

```python
mytable
```

```python
ForGumpUserRate = mytable['Forrest Gump (1994)']
```

```python
ForGumpRecs = mytable.corrwith(ForGumpUserRate)
ForGumpRecs.dropna(inplace=True)
```

```python
len(ForGumpRecs)
```

```python
ForGumpRecs = pandas.DataFrame(ForGumpRecs,columns=['correlation'])
```

```python
ForGumpRecs = ForGumpRecs.join(mydata['count'],how='inner')
```

```python
ForGumpRecs
```

```python
rateLIMIT = 20
```

```python
print(ForGumpRecs[ForGumpRecs['count']>rateLIMIT].sort_values('correlation',ascending=False).head(16))
```
